```python
def clean(text):
  text=str(text).lower()
  text=re.sub(r'[^a-zA-Z ]', ' ', text)
  words=text.split()
  words=[w for w in words if w not in stop_words]
  return " ".join(words)
```

```python
def extract_skills(text,role=TARGET_ROLE):
    skills = []
    for skill in SKILLS_BY_CATEGORY[role]:
        if skill in text:
            skills.append(skill)
    return skills
```

```python
data = df[df['Category'] == TARGET_ROLE].copy().reset_index(drop=True)
data['Resume_str'].apply(clean)
```

```python
data['skills']=df['Resume_str'].apply(extract_skills)
```

```python
job_clean  = clean(JOB_DESCRIPTION)
job_skills=extract_skills(job_clean)
```

```python
job_skills
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vec=TfidfVectorizer()
all_text = data['Resume_str'].tolist() + [job_clean]
mat=vec.fit_transform(all_text)
X=mat[:-1]
y=mat[-1]
data['score']=cosine_similarity(X,y).flatten()
```

```python
data['score'].head()
```

```python
def skill_gap(candidate_skills):
  matched = list(set(job_skills) & set(candidate_skills))
  missing = list(set(job_skills) - set(candidate_skills))
  return matched, missing
```

```python
data['matched_Skills'],data['missing_Skills']=zip(*data['skills'].apply(skill_gap))
```

```python
data['skill_match_pct'] = data['matched_Skills'].apply(
    lambda x: round(len(x) / len(job_skills) * 100, 1) if job_skills else 0
)
```

```python
data['final_score'] = (
    0.30 * data['score'] +
    0.70 * (data['skill_match_pct'] / 100)
)
```

```python
data['Rank'] = data['final_score'].rank(ascending=False, method='dense').astype(int)
```

```python
data=data.sort_values(by='Rank',ascending=True)
```
